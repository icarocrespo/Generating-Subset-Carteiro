```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
x = pd.Series([12645.656565, 12645.989755, 12645.434345])
```

```python
y = pd.Series([42968.331100, 42918.398400, 42765.459070])
```

```python
f = open("bairros.txt","r")
l_bairros = f.read().split()
f.close()
```

```python
cx = []
cy = []
for x in range(0,len(l_bairros),3):
    cx.append(l_bairros[x+1])
    cy.append(l_bairros[x+2])

plt.plot(cx, cy,'o')
plt.title("Bairros de Alegrete")
plt.show()
```
