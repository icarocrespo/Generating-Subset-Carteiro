## Plotagem das rotas: bairros de Alegrete ##

```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
x = pd.Series([29.8089074, 29.8048719, 29.782602, 29.773959, 29.795815, 29.7841203])
```

```python
y = pd.Series([55.8064986, 55.8043767, 55.8074572, 55.7932329, 55.7539836, 55.7906762])
```

```python
f = open("bairros16.txt","r")
l_bairros = f.read().split()
f.close()
```

```python
cx = []
cy = []
for x in range(0,len(l_bairros),3):
    cx.append(l_bairros[x+1])
    cy.append(l_bairros[x+2])

plt.plot(cx, cy,'o')
plt.title("Bairros de Alegrete")
plt.show()
```
